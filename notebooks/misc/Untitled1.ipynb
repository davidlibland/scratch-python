```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
k = 3
n= 7
x = range(1, n)
y = np.array([1/k if i <=k else 0 for i in x ])
```

```python
plt.figure(figsize=(2,2))
plt.scatter(x = [k]*(k), y = range(1,k+1))
plt.xlim(1,n)
plt.ylim(1,n)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
fig = plt.figure()
ax = fig.gca(projection='3d')

# Make data.
X = np.arange(1, n, .25)
Y = np.arange(1, n, .25)
X, Y = np.meshgrid(X, Y)
Z = 0.5*((1/X*np.heaviside(X-Y,.5))+(1/Y * np.heaviside(Y-X,.5)))

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
```
