```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
%matplotlib inline
```

```python
from scipy.special import expit
```

```python
x = np.linspace(-5,5,50)
y = np.linspace(-5,5,50)
X, Y = np.meshgrid(x, y)
```

```python
Z1 = expit(X)*expit(Y)
Z2 = expit(X+Y)
Z3 = (expit(X)+expit(Y))*0.5
```

```python
plt.pcolor(X,Y,Z1)
```

```python
plt.pcolor(X,Y,Z2)
```

```python
plt.pcolor(X,Y,Z3)
```

```python
plt.pcolor(X,Y,np.sqrt(Z1)+Z2-Z3)
```

```python
def generate_gp(kern, x_min, x_max, n):
    xs = np.linspace(x_min, x_max, n)
    cov_mat = np.array([
        [kern(x_i, x_j) for x_j in xs]
        for x_i in xs
    ])
    return xs, scipy.stats.multivariate_normal.rvs(cov=cov_mat)
```

```python
xs, ys = generate_gp(lambda x,y: min(x,y), 0, 1, 300)
plt.plot(xs,ys)
```

```python
xs, ys = generate_gp(lambda x,y: min(x,y), 0, .1, 300)
plt.plot(xs,ys)
```

```python
xs, ys = generate_gp(lambda x,y: min(x,y), 0, .01, 300)
plt.plot(xs,ys)
```

```python
xs, ys = generate_gp(lambda x,y: x*y, 0, 1, 300)
plt.plot(xs,ys)
```
