```python
from scipy.stats import beta
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 100)
```

```python
rv = beta(32, 34)
plt.plot(x, rv.pdf(x), 'k-', lw=2, label='frozen pdf')
```

```python
rv = beta(37, 3)
plt.plot(x, rv.pdf(x), 'k-', lw=2, label='frozen pdf')
```
